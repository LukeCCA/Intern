```python
def show_distribute(file_path):
    df = pd.read_csv(file_path, sep='\t')
    labels = sorted(set(df.labels.tolist()))
    num_labels = len(labels)

    count = df.groupby('labels').count()["texts"]
    pd.set_option('display.max_rows', None)
#     print(count)
    plt.bar(list(range(0,num_labels)), count)
    plt.show()
    return
```

### tags

```python
tags = {}
with open(f"{FOLDER_PATH}/tag.txt", 'r') as fp:
    for line in fp:
        line = re.sub('\s+', '', line)
        line = line.split('=')
        replace_word = [text for text in line[1].split(',') if text != '']
        tags[line[0]] = replace_word
# tags
```

```python
for mode in ["train", "test"]:
    df = pd.read_csv(f"data/IVR/{mode}.tsv", sep='\t')
    texts = df["texts"].tolist()
    tf_list = []

# tags = {"a": ["存摺", "金融卡", "卡", "卡片"]}

# texts = ["存摺金融卡遺失"]
    for text in texts:
        tf = [0 for _ in range(len(tags))]
        i = 0
        for key, value in tags.items():
            value = sorted(value, key=len, reverse=True)
#             print(value)
    #         value = [v for v in value if len(v)==len(text)]

    #         if value == []:
    #             i += 1
    #             continue

            big_regex = re.compile('|'.join(map(re.escape, value)))
            temp = big_regex.findall(text)
            
            tf[i] += len(temp)
            i += 1
# print((tf))
        tf_list += [tf]
    df["tf"] = tf_list
    df.to_csv(f"data/IVR/{mode}.tsv", sep='\t', index=False)
```

### class label

```python
labels = []
vectors = []
with open(f"{FOLDER_PATH}/class_label.txt") as fp:
    for line in fp:
        vector = [0] * len(tags)
        line = re.sub('\s+', '', line)
        if line == '':
            continue
        line = line.split('=')
        label = line[0]
        labels += [label[label.find('_')+1 : label.rfind('_')]]
        feature = line[1].split(',')[1:]
        for f in feature:
            try:
                idx = list(tags.keys()).index(f)
                vector[idx] = 1
            except:
                pass
        vectors += [vector]
df = pd.DataFrame({"labels":labels, "vectors":vectors})
df.head()
```

```python
df.to_csv("data/IVR/class_label.tsv", sep='\t', index=False)
```

## SMP2017

```python
FOLDER_PATH += "SMP2017"
```

```python
def file_iter(mode):
    file_name_ls = [file_name for file_name in os.listdir(f"{FOLDER_PATH}/{mode}") if file_name.find(".txt")!=-1]
    for name in file_name_ls:
        with open(f"{FOLDER_PATH}/{mode}/{name}") as f:
            yield f.readlines(), name[name.find('_')+1:-4]
```

```python
def txt_2_tsv(file_path, mode):
    texts = []
    labels = []
    for ls, label in tqdm(file_iter(mode)):
        for t in ls:
            texts += [t[:-1]] # -1 for '\n'
        labels += [label] * len(ls)
      
    df = pd.DataFrame({"texts":texts, "labels":labels})
    df.to_csv(file_path, sep='\t', index=False)
    print(df.head())
    
    return 
```

```python
for mode in ["train", "develop", "test"]:
    file_path = f"data/SMP2017/{mode}.tsv"
    txt_2_tsv(file_path, mode)
    show_distribute(file_path)
```

## SMP2018

```python
def json_2_tsv(input_file, output_file):
    converter = opencc.OpenCC('s2t.json')
    with open(input_file, "r") as fp:
        json_data = json.load(fp)
    texts = []
    labels = []
    for i in range(len(json_data)):
        texts += [converter.convert(json_data[str(i)]["query"])]
        labels += [json_data[str(i)]["label"]]
    df = pd.DataFrame({"texts": texts, "labels":labels})
    df.to_csv(output_file, sep='\t', index=False)
    print(df.head())
    return
```

```python
json_2_tsv("extend_dataset/SMP2018/train.json", "data/SMP2018/train.tsv")
show_distribute("data/SMP2018/train.tsv")
```
